## 1. Impedancia de entrada y valores normalizados

Simplifico el circuito para centrarme en las impedancias:

Relación de tensiones en los nodos de los operacionales:$$V_A = V_C = V_E = V_x$$

Corriente de entrada:
$$I_{in} = \frac{V_A - V_B}{Z_1} = \frac{V_x - V_B}{Z_1}$$

Ecuaciones de corriente:

En el nodo $C$:

$$I_{z2} = I_{z3}$$

$$\frac{V_{z2}}{Z_2} = \frac{V_{z3}}{Z_3}$$

$$\frac{V_B - V_C}{Z_2} = \frac{V_C - V_D}{Z_3}$$

$$\frac{V_B - V_x}{Z_2} = \frac{V_x - V_D}{Z_3}$$

$$V_B - V_x = \frac{Z_2}{Z_3} (V_x - V_D)$$


En el nodo $E$:
$$\quad I_{R4} = I_{R5}$$

$$\frac{V_{R4}}{R_4} = \frac{V_{R5}}{R_5}$$

$$\frac{V_D - V_E}{R_4} = \frac{V_E}{R_5}$$

$$\frac{V_x - V_D}{R_4} = \frac{-V_x}{R_5}$$

$$V_x - V_D = -\frac{R_4}{R_5} V_x$$

Reemplazando en $V_B - V_x = \frac{Z_2}{Z_3} (V_x - V_D)$:

$$V_B - V_x = -\frac{Z_2}{Z_3} \cdot \frac{R_4}{R_5} \, V_x$$

Teniendo en cuenta la $I_{in}$ antes definida:

$$ -I_{in} Z_1 = - \frac{Z_2}{Z_3} \frac{R_4}{R_5} V_x $$

$$\frac{V_x}{I_{in}} = \frac{Z_1 Z_3 R_5}{Z_2 R_4} = \frac{Y_2 R_5}{Y_1 Y_3 R_4} $$

Dado que se busca el comportamiento de un inductor unitario:
$$ \left. \begin{aligned} Z_{in} = s &= \frac{Y_2 R_5}{Y_1 Y_3 R_4} \\ Y_C &= sC \end{aligned} \right\} \Rightarrow \begin{aligned} Y_2 &= sC_2 \\ C_2 \cdot R_5 &= \frac{R_4}{R_1 R_3} \end{aligned} $$

Estando normalizados, los valores y unidades quedan así:
$$Y_1 = Y_3 = 1S$$
$$R_4 = R_5 = 1Ω $$
$$ Y_2 = 1F$$



## 2. Función de Transferencia y Desnormalización

$$ \frac{V_2}{V_1} = \frac{V_2}{V_x} \frac{V_x}{V_1} $$

Siendo $V_2 = V_D$

Anteriormente planteado por el nodo $E$:

$$ V_x - V_D = - \frac{R_4}{R_5} V_x $$

$$+V_2 = + \frac{R_4}{R_5} V_x + V_x $$

$$\frac{V_2}{V_x} = \frac{R_4}{R_5} + 1 $$

$$ V_x = V_1  \frac{R}{R + (\frac{1}{sC} \parallel sL)} $$
$$ \frac{V_x}{V_1} = \frac{(\frac{1}{sC} \parallel sL)}{R + (\frac{1}{sC} \parallel sL)} $$

$$ \frac{1}{sC} \parallel sL = \left[ \left( \frac{1}{sC} \right)^{-1} + (sL)^{-1} \right]^{-1} = \left( sC + \frac{1}{sL} \right)^{-1} = \left[ \frac{sL}{sL} \left( sC + \frac{1}{sL} \right) \right]^{-1} = \left[ \frac{1}{sL} (s^2 LC + 1) \right]^{-1} = \frac{sL}{s^2 LC + 1} $$

$$ \frac{V_x}{V_1} = \frac{sL}{(s^2 LC + 1)} \frac{1}{(R + \frac{sL}{s^2 LC + 1})} = \frac{sL}{s^2 RLC + R + sL} = \frac{s}{RC (s^2 + s \frac{1}{RC} + \frac{1}{LC})} $$


Función transferencia:
$$ \frac{V_2}{V_x} \frac{V_x}{V_1} = \left( \frac{R_4}{R_5} + 1 \right) \frac{s}{RC \left( s^2 + s \frac{1}{RC} + \frac{1}{LC} \right)} $$

$$ \omega_0 = \sqrt{\frac{1}{LC}} $$
$$ \frac{\omega_0}{Q} = \frac{1}{RC} \implies Q = R \sqrt{\frac{C}{L}} $$

$$ \left. \begin{aligned} \omega_{0n} &= 1 \text{ rad/s} \\ L_n &= 1 \text{ H} \end{aligned} \right\} \implies 1 \text{ rad/s} = \frac{1}{\sqrt{1 \text{ H} \cdot C}} \longrightarrow C_n = 1 \text{ F} $$

$$ Q_n = 20 = R_n \sqrt{\frac{1 \text{ F}}{1 \text{ H}}} \longrightarrow R_n = 20 \ \Omega $$

$$ \Omega_\omega = 2\pi f_0 = 62\,831,85 $$

$$ C_n = C \cdot \Omega_\omega \cdot \Omega_z \longrightarrow \Omega_z = \frac{C_n}{C \cdot \Omega_\omega} = 1591,55 $$

$$ L = \frac{L_n \cdot \Omega_z}{\Omega_\omega} = 25,33 \text{ mH} $$

$$ R = R_n \cdot \Omega_z = 31,8 \text{ k}\Omega $$

$$ 25,33 \text{ mH} = \frac{10\text{nF} \cdot R_1 \cdot R_3 \cdot R_5}{R_4}$$

$$ 2533000 = \frac{R_1 \cdot R_3 \cdot R_5}{R_4} $$

## 3. Transferencia desnormalizada

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

R = 31.8e3
C = 10e-9
L = 25.33e-3

R4 = 1.6e3
R5 = 180
K = 1 + (R4 / R5)

num = [K * (1 / (R * C)), 0]

den = [1, 1 / (R * C), 1 / (L * C)]

filtro_pb = signal.TransferFunction(num, den)

f_hz = np.logspace(3, 5, 1000)
w_rad = 2 * np.pi * f_hz

w, mag, phase = signal.bode(filtro_pb, w_rad)

plt.figure(figsize=(10, 6))

plt.semilogx(f_hz, mag, label='Respuesta del Filtro', color='b', linewidth=2)

plt.axvline(10e3, color='r', linestyle='--', label='$f_0 = 10$ kHz')
plt.axhline(20 * np.log10(K), color='g', linestyle=':', label=f'Ganancia Max $\\approx$ {20 * np.log10(K):.1f} dB')

plt.title('Diagrama de Bode', fontsize=14)
plt.xlabel('Frecuencia [Hz]', fontsize=12)
plt.ylabel('Magnitud [dB]', fontsize=12)
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend(loc='upper right')

plt.show()

f0_simulada = f_hz[np.argmax(mag)]
print(f"--- Verificación ---")
print(f"Frecuencia central simulada: {f0_simulada:.2f} Hz")
print(f"Ganancia máxima simulada: {np.max(mag):.2f} dB")
```
