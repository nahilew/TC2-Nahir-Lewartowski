#Red T

Analíticamente:

$$Z = \begin{pmatrix} Z_a + Z_c & Z_c \\ Z_c & Z_b + Z_c \end{pmatrix}$$

$$\Delta_z = Z_a Z_b + Z_a Z_c + Z_b Z_c$$

Transformaciones:

$$Y = \begin{pmatrix} Z_{22} & -Z_{12} \\ -Z_{21} & Z_{11} \end{pmatrix} \frac{1}{\Delta_z} = \begin{pmatrix} Z_b + Z_c & -Z_c \\ -Z_c & Z_a + Z_c \end{pmatrix} \frac{1}{Z_a Z_b + Z_a Z_c + Z_b Z_c}$$

$$T = \begin{pmatrix} Z_{11} & \Delta_z \\ 1 & Z_{22} \end{pmatrix} \frac{1}{Z_{21}} = \begin{pmatrix} Z_a + Z_c & Z_a Z_b + Z_a Z_c + Z_b Z_c \\ 1 & Z_b + Z_c \end{pmatrix} \frac{1}{Z_c}$$

#Red $π$

Analíticamente:

$$Y  = \begin{pmatrix} Y_A + Y_C & -Y_C \\ -Y_C & Y_B + Y_C \end{pmatrix}$$

$$\Delta_Y = Y_A Y_B + Y_A Y_C + Y_B Y_C$$

Transformaciones:

$$Z = \begin{pmatrix} Y_{22} & -Y_{12} \\ -Y_{21} & Y_{11} \end{pmatrix} \frac{1}{\Delta_Y} = \begin{pmatrix} Y_B + Y_C & Y_C \\ Y_C & Y_A + Y_C \end{pmatrix} \frac{1}{Y_A Y_B + Y_A Y_C + Y_B Y_C}$$

$$T = \begin{pmatrix} -Y_{22} & -1 \\ -\Delta_Y & -Y_{11} \end{pmatrix} \frac{1}{Y_{21}} = \begin{pmatrix} Y_B + Y_C & 1 \\ Y_A Y_B + Y_A Y_C + Y_B Y_C & Y_A + Y_C \end{pmatrix} \frac{1}{Y_C}$$

#Red Lattice

Analíticamente:

$$Z = \begin{pmatrix} \frac{Z_b + Z_a}{2} & \frac{Z_b - Z_a}{2} \\ \frac{Z_b - Z_a}{2} & \frac{Z_b + Z_a}{2} \end{pmatrix}$$

$$\Delta_Z = Z_a Z_b$$

Transformaciones:

$$Y = \begin{pmatrix} Z_{22} & -Z_{12} \\ -Z_{21} & Z_{11} \end{pmatrix} \frac{1}{\Delta_Z} = \begin{pmatrix} \frac{Z_b + Z_a}{2} & \frac{Z_a - Z_b}{2} \\ \frac{Z_a - Z_b}{2} & \frac{Z_b + Z_a}{2} \end{pmatrix} \frac{1}{Z_a Z_b}$$

$$T = \begin{pmatrix} Z_{11} & \Delta_Z \\ 1 & Z_{22} \end{pmatrix} \frac{1}{Z_{21}} = \begin{pmatrix} Z_b + Z_a & 2 Z_a Z_b \\ 2 & Z_b + Z_a \end{pmatrix} \frac{1}{Z_b - Z_a}$$
